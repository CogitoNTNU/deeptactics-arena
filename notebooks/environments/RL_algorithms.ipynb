# RL algorithms

Forskjellige algoritmer for reinforcement learning.

## HER

HER står for Hindsight Experience Replay.

Algoritme som bruker andre algoritmer.

Funker best for miljøer hvor man for lite reward og egentlig kun rewards på perfekt utførelse.

Later som at man klarer andre mål selv om det ikke har gått så bra for modellen nå.
For eksempel hvis agenten spiller fotball og skyter langt til høyre for mål får den null reward.
Det HER gjør er at den later som at målet er til høyre og med det lærer modellen at hvis skuddet hadde vært mer til venstre hadde det blitt mål, og hadde fått reward.

Den bruker en slags replay buffer for å få til dette, og bruker altså andre algoritmer.

## A2C

A2C står for Advantage Acor-Critic.

Algoritmen bruker fordelene med både policy-based og value-based metoder.

Består av to ulike deler, en actor og en critic.
